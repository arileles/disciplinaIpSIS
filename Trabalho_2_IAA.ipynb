```python
# Atividade 2.a
import pandas as pd
data = pd.read_csv("/content/data.csv", header=None)
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go #2f
from mpl_toolkits.mplot3d import Axes3D #2g
from sklearn.linear_model import LinearRegression #2i
from mpl_toolkits.mplot3d import Axes3D
import statsmodels.api as sm
from scipy.interpolate import griddata

print("-------------------------------------------------------------------------------------------------------")

# Atividade 2.b
df = pd.DataFrame(data)

media = df[2].describe().mean().round(2)

minimo = df[2].describe().min()

quartos = df.loc[df[2].idxmax(), 1]

print('Média preço casa: ',media,',  preço menor casa: ', minimo, ',  quantos quartos tem a casa mais cara: ',quartos)

print("-------------------------------------------------------------------------------------------------------")

# Atividade 2.c
# Matriz X para variáveis independentes
X = df[[0, 1]].values

# Vetor y para a variável dependente
y = df[2].values

print("Matriz X:")
print(X)

print("\nVetor y:")
print(y)

print("-------------------------------------------------------------------------------------------------------")

# Atividade 2.d

def correlacao(x, y):

    # Calcula as médias
    x_mean = np.mean(x)
    y_mean = np.mean(y)

    # Calcula o coeficiente de correlação
    numerator = np.sum((x - x_mean) * (y - y_mean))
    denominatorX = np.sum((x - x_mean) ** 2)
    denominatorY = np.sum((y - y_mean) ** 2)

    raiz_quadrada = np.sqrt(denominatorX*denominatorY)

    #Calcula correlacao
    r = numerator / raiz_quadrada

    return r


def regressao(x, y):

    # Calcula as médias
    x_mean = np.mean(x)
    y_mean = np.mean(y)

    # Calcula os coeficientes da regressão
    numerator = np.sum((x - x_mean) * (y - y_mean))
    denominatorX = np.sum((x - x_mean) ** 2)

    # Coeficiente angular (β1)
    beta1 = numerator / denominatorX

    # Intercepto (β0)
    beta0 = y_mean - beta1 * x_mean

    return (beta0, beta1)


r = correlacao(df[0], df[2])
print("Coeficiente de Correlação de Tamanho da casa e Preço:", r)
# Calculando regressão
beta02, beta12 = regressao(df[0], df[2])
print("Intercepto (β0):", beta02)
print("Coeficiente Angular (β1):", beta12)

plt.scatter(df[0], df[2], label='Dados')
plt.plot(df[0], beta02 + beta12 * df[0], color='red', label='Regressão Linear')
plt.xlabel('X')
plt.ylabel('Y')
plt.title(f'Regressão Linear - Exemplo 2\nCorrelação: {r:.5f}')
plt.legend()
plt.show()

print("\n")

r = correlacao(df[1], df[2])
print("Coeficiente de Correlação de Número de quartos e Preço:", r)
# Calculando regressão
beta03, beta13 = regressao(df[1], df[2])
print("Intercepto (β0):", beta03)
print("Coeficiente Angular (β1):", beta13)

plt.scatter(df[1], df[2], label='Dados')
plt.plot(df[1], beta03 + beta13 * df[1], color='red', label='Regressão Linear')
plt.xlabel('X')
plt.ylabel('Y')
plt.title(f'Regressão Linear - Exemplo 2\nCorrelação: {r:.5f}')
plt.legend()
plt.show()

print("-------------------------------------------------------------------------------------------------------")
print("2e")
# Atividade 2.e

# Definindo as variáveis independentes e dependentes
X = df[[0, 1]]
y = df[2]

# Adicionando uma constante ao modelo (intercepto)
X = sm.add_constant(X)

# Criando o modelo de regressão
modelo = sm.OLS(y, X).fit()

# Resumo do modelo
print(modelo.summary())

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
ax.scatter(df[0], df[1], df[2], color='b', label='Dados')

# Definindo os rótulos dos eixos
ax.set_xlabel('Tamanho da Casa (sq ft)')
ax.set_ylabel('Número de Quartos')
ax.set_zlabel('Preço da Casa ($)')
ax.set_title('Gráfico de Dispersão 3D')

# Mostrando a legenda
ax.legend()

plt.show()
print("-------------------------------------------------------------------------------------------------------")
print("2f")


# Definindo os eixos
x = df[0].values
y = df[1].values
z = df[2].values
# Criando uma grade para a superfície
xi = np.linspace(x.min(), x.max(), 100)
yi = np.linspace(y.min(), y.max(), 100)
x_surf, y_surf = np.meshgrid(xi, yi)

# Interpolando os dados
z_surf = griddata((x, y), z, (x_surf, y_surf), method='cubic')

# Criando o gráfico
fig = go.Figure()

# Adicionando os dados reais
fig.add_trace(go.Scatter3d(x=x, y=y, z=z, mode='markers', name='Dados Reais',
                             marker=dict(size=5, color='blue')))

# Adicionando a superfície interpolada
fig.add_trace(go.Surface(x=x_surf, y=y_surf, z=z_surf, name='Superfície Interpolada',
                         colorscale='Viridis', opacity=0.8, showscale=True))

# Atualizando layout
fig.update_layout(scene=dict(xaxis_title='Tamanho da Casa',
                              yaxis_title='Número de Quartos',
                              zaxis_title='Preço'),
                  title='Gráfico de Dispersão com Superfície Interpolada em 3D')

fig.show()

print("-------------------------------------------------------------------------------------------------------")
print("2g")

x = df[0].values  # Tamanho da casa
y = df[1].values  # Número de quartos
z = df[2].values  # Preço

# Calculando os coeficientes de correlação
cor_tamanho_preco = np.corrcoef(df[0], df[2])[0, 1]  # Tamanho da casa e Preço
cor_quartos_preco = np.corrcoef(df[1], df[2])[0, 1]  # Número de quartos e Preço

# Criando uma grade para a superfície de interpolação
xi = np.linspace(x.min(), x.max(), 100)
yi = np.linspace(y.min(), y.max(), 100)
x_surf, y_surf = np.meshgrid(xi, yi)

# Interpolando os dados
z_surf = griddata((x, y), z, (x_surf, y_surf), method='cubic')

# Criando o gráfico
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Adicionando os dados reais
ax.scatter(x, y, z, color='blue', label='Dados Reais', alpha=0.6)

# Plotando a superfície interpolada
surf = ax.plot_surface(x_surf, y_surf, z_surf, cmap='viridis', alpha=0.8)

# Adicionando a barra de cores
fig.colorbar(surf, shrink=0.5, aspect=5)

# Definindo os rótulos dos eixos
ax.set_xlabel('Tamanho da Casa')
ax.set_ylabel('Número de Quartos')
ax.set_zlabel('Preço')

# Definindo o título
ax.set_title('Gráfico de Dispersão com Superfície Interpolada em 3D')

# Adicionando os coeficientes de correlação ao gráfico
ax.text2D(0.05, 0.95, f'Coeficiente de Correlação (Tamanho da casa e Preço): {cor_tamanho_preco:.2f}', transform=ax.transAxes, fontsize=10)
ax.text2D(0.05, 0.90, f'Coeficiente de Correlação (Número de quartos e Preço): {cor_quartos_preco:.2f}', transform=ax.transAxes, fontsize=10)

# Exibindo a legenda
ax.legend(loc='upper left', bbox_to_anchor=(0.05, 0.8))

plt.show()

print("-------------------------------------------------------------------------------------------------------")
print("2h")

#Atividade 2 H
# Regressão múltipla: calculando os coeficientes
def gauss_jordan_inverse(matrix):
    n = matrix.shape[0]
    # Cria uma matriz aumentada com a matriz original e a matriz identidade
    augmented = np.hstack((matrix, np.eye(n)))

    for i in range(n):
        # Normaliza a linha i
        augmented[i] = augmented[i] / augmented[i, i]
        for j in range(n):
            if i != j:
                augmented[j] = augmented[j] - augmented[i] * augmented[j, i]

    # Retorna a parte da matriz aumentada que corresponde à inversa
    return augmented[:, n:]

def regressao_multipla(X, y):
    # Adiciona uma coluna de 1s para o intercepto
    X_b = np.c_[np.ones((X.shape[0], 1)), X]  # coluna de 1s


    XtX = X_b.T @ X_b


    XtY = X_b.T @ y


    XtX_inv = gauss_jordan_inverse(XtX)

    # Calcula os coeficientes
    theta_best = XtX_inv @ XtY

    return theta_best

# Variáveis independentes (Tamanho da casa e Número de quartos)
X = df[[0, 1]].values  # Tamanho da casa e Número de quartos
y = df[2].values  # Preço da casa

coeficientes = regressao_multipla(X, y)

beta0 = coeficientes[0]
beta1 = coeficientes[1]
beta2 = coeficientes[2]

# Tamanho da casa e variações do número de quartos
tamanho_casa = 1650
quartos_variacoes = [2, 3, 4]

# Calculando o preço para a casa com 1650 de tamanho
for quartos in quartos_variacoes:
    preco = beta0 + beta1 * tamanho_casa + beta2 * quartos  # Usando todos os coeficientes
    print(f'Preço da casa com tamanho {tamanho_casa} e {quartos} quartos: R$ {preco:.2f}')

print("\nAo aumentar o número de quartos, o preço tende a aumentar, pois isso geralmente indica uma casa mais desejável.")
print("Se diminuir o número de quartos, o preço tende a cair, pois isso pode tornar a casa menos atrativa para compradores.")

print("-------------------------------------------------------------------------------------------------------")
print("2i")

#Atividade 2.i
# Criando o modelo de regressão linear múltipla com scikit-learn
modelo = LinearRegression()

modelo.fit(X, y)

# Coeficientes
coeficientes = modelo.coef_
intercepto = modelo.intercept_

# Fazendo previsões
y_pred = modelo.predict(X)

# Comparando os resultados
print("Coeficientes usando scikit-learn:")
print(f"Coeficiente angular (β1) para Tamanho da casa: {coeficientes[0]:.5f}")
print(f"Coeficiente angular (β2) para Número de quartos: {coeficientes[1]:.5f}")
print(f"Intercepto (β0): {intercepto:.5f}")

# Exibindo algumas previsões
print("\nPrevisões de preço usando scikit-learn:")
for i in range(5):
    print(f'Preço da casa com Tamanho {X[i][0]} e {X[i][1]} quartos: R$ {y_pred[i]:.2f}')
```
